## Obtenção das tabelas e dados

```python
pasta_dos_csv = '1-datasets/'

for nome_do_arquivo in os.listdir(pasta_dos_csv):
    if nome_do_arquivo.endswith('.csv'):
        caminho_do_arquivo = os.path.join(pasta_dos_csv, nome_do_arquivo)
        variavel_nome = nome_do_arquivo.replace('.csv', '').replace('-','_').replace(' ', '_')
        try:
            globals()[f'dataset_{variavel_nome}'] = pd.read_csv(caminho_do_arquivo, sep=';')
        except Exception as e:
            print(f'Falha no carregamento no arquivo {nome_do_arquivo}: {e}')
```


**O que está bem feito:**
- `os.path.join` — portável entre Windows, Mac e Linux
- `.endswith('.csv')` — não tenta ler outros tipos de arquivo
- `replace('-', '_')` e `replace(' ', '_')` — garante nomes de variável válidos em Python
- `try/except` — se um arquivo estiver corrompido, o loop continua

# Análise Exploratória de Dados (EDA)

```python
def eda_report(df, nome='DataFrame'):
    print(f"{'='*40}")
    print(f"EDA Report: {nome}")
    print(f"{'='*40}")
    # Colunas e Linhas
    print(f"\nColunas: {df.shape[1]}")
    print(f"Linhas:  {df.shape[0]}")
    # Dtypes
    print(f"\nDtypes:")
    print(f"  str:   {len(df.select_dtypes(include=['str','object']).columns)}")
    print(f"  int:   {len(df.select_dtypes(include='int').columns)}")
    print(f"  float: {len(df.select_dtypes(include='float').columns)}")
    print(f"  data:  {len(df.select_dtypes(include='datetime').columns)}")
    # Non-Null
    print(f"\nNon-Null:")
    for col in df.columns:
        print(f"  {col}: {df[col].notnull().sum()}")
    # Unique
    print(f"\nUnique:")
    for col in df.columns:
        print(f"  {col}: {df[col].nunique()}")
    # Duplicatas
    print(f"\nDados Duplicados: {df.duplicated().sum()}")
    print(f"{'='*40}")
```
